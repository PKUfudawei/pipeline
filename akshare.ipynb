```python
import akshare
import awkward as ak
import os
```

```python
df_tencent = akshare.stock_hk_hist(symbol="00700", period="daily", start_date="20120101", end_date="20121231", adjust='qfq')
df_tencent
```

```python
df_tencent = df_tencent.rename(columns={'日期':'Date', '开盘': 'Open', '收盘': 'Close', '最高': 'High', '最低': 'Low', '成交量': 'Volume'})
df_tencent=df_tencent.loc[:, ['Date', 'Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
if not os.path.exists('./data/'):
    os.makedirs('./data')
with open('./data/test_stock.txt', 'w', encoding='utf-8') as f:
    df_tencent.to_csv(f, index=False)
```
